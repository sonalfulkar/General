# REGEX

```python
import re
# Matches only start of the String
result = re.match(r'AV', 'AV Aanalytics Vidya AV')
print(result)
```

```python
print(result.group(0))
```

```python
result = re.match(r'Analytics', 'AV Analytics Vidya AV')
print(result)
```

```python
result = re.match(r'AV', 'AV Analytics Vidya AV')
print(result.start(), result.end())
```

```python
# Match anywhere in the String
result = re.search(r'Vidya', 'AV Analaytics Vidya AV')
print(result.group(0))
```

```python
# search all occurences from a string
result = re.findall(r'AV', 'AV Analytics Vidya AV, AV again')
print(result)
```

```python
# Split the string by given pattern
result = re.split(r'y', 'Analytics go to type y again y')
print(result)
```

```python
#maxsplit - to split only given number of times
result = re.split(r'y', 'Analytics Vidya', maxsplit=1)
print(result)
```

```python
# sub-string search and replace
result = re.sub(r'India', 'the world', 'Analytics vidya is famous in India')
print(result)
```

```python
# compiling a pattern, to reuse
pattern = re.compile(r'AV')
result = pattern.match('AV analytics Vidya')
result2 = pattern.search('Analytics AV')
result3 = pattern.findall('AV again AV')
result4 = pattern.sub('GV', 'AV Again AV')
result5 = pattern.split('AV again AV 2')
print('1--', result.group(0), '2--', result2.group(0), '3--', result3, '4--', result4, '5---', result5)
```

```python
# find first word from given line
import re
line = 'AV is largest Analytics community of India'
result1 = re.match(r'.', line) # any non-newline chars
result2 = re.findall(r'\w', line) # any word chars (alphanumeric, no spaces)
result3 = re.findall(r'\w*', line) # 0 or more words
result4 = re.findall(r'\w+', line) # 1 or more workds, spaces excluded
result5 = re.findall(r'^\w+', line) # get word from start of string
print('1st attempt - ', result1, '\n2nd attempt', result2, '\n3rd attempt', result3, '\n4th attempt', result4, '\n5th attempt', result5)
```

```python
# return first 2 characters of each word
line = 'AV is largest Analytics community of India'
result1 = re.findall(r'\w\w', line)
result2 = re.findall(r'\b\w.', line)
print(result1, '\n',result2)
```

```python
# return domain type of email ID
line = 'abc.test@gmail.com, xyz@test.in, test.first@analyticsvidhya.com, first.test@rest.biz'
result1 = re.findall(r'@\w+', line) # fetch all words after @
result2 = re.findall(r'@\w+\.\w+', line) # fetch words after . as well
result3 = re.findall(r'@\w+\.(\w+)', line) # only fetch words inside ()
print(result1, '\n', result2, '\n', result3)
```

```python
# return date from given string
line = 'Sonal 34-3456 12-05-2007, XYZ 56-4532 11-11-2011, ABC 67-8945 12-01-2009'
result1 = re.findall(r'\d\d-\d\d-\d\d\d\d', line)
result2 = re.findall(r'\d{2}-\d{2}-\d{4}', line)
result3 = re.findall(r'\d{2}-\d{2}-(\d{4})', line) # if only year
print(result1, '\n', result2, '\n', result3)
```

```python
# return all words starting with vowel
line = 'AV is largest Analytics community of India'
result1 = re.findall(r'\w+', line)
result2 = re.findall(r'[aeiouAEIOU]\w+', line)
result3 = re.findall(r'\b[aeiouAEIOU]\w+', line)
result4 = re.findall(r'\b[^aeiouAEIOU]\w+', line) # non vowel starting words
result5 = re.findall(r'\b[^aeiouAEIOU ]\w+', line)
print(result1, '\n', result2, '\n', result3, '\n', result4, '\n', result5)
```

```python
# Validate phone numbers - 10 digit starting with 8 or 9
phones = ['9999999999','999999-999','99999x9999']
for phone in phones:
    if re.match(r'[8-9]{1}[0-9]{9}', phone) and len(phone) == 10:
        print(phone, "Match")
    else:
        print(phone, 'Not match')
```

```python
# split a string with multiple delimiters
line = 'asdf fjdk;afed,fjek,asdf,foo' # String has multiple delimiters (";",","," ").
result1 = re.split(r'[;,\s]', line)
result2 = re.sub(r'[;,\s]', ' ', line) # replace all delimiters by single space delimiter
print(result1, '\n', result2)
```
